```python
#Data Cleaning
df.isnull().sum()
```

```python
#Data Processing
x=df.iloc[:,[3,4]].values
```

```python
print(x)
```

```python
df.nunique()
```

```python
#Model Fitting
from sklearn.cluster import KMeans
WCSS=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)
```

```python
kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
y=kmeans.fit_predict(x)
print(y)
```

```python
sns.set()
plt.plot(range(1,11),WCSS)
plt.title('the elbow point graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#production of crop
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1] ,s=50, c='green', label='cluster 1')
plt.scatter(x[y==1,0],x[y==1,1] ,s=50, c='red', label='cluster 2')
plt.scatter(x[y==2,0],x[y==2,1] ,s=50, c='blue', label='cluster 3')
plt.scatter(x[y==3,0],x[y==3,1] ,s=50, c='yellow', label='cluster 4')
plt.scatter(x[y==4,0],x[y==4,1] ,s=50, c='violet', label='cluster 5')
#plt.scatter(x[y==5,0],x[y==5,1] ,s=50, c='green', label='cluster 1')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label='Centroids')
plt.title('Crop Prediction')
plt.ylabel('Production')
plt.xlabel('Yield')
plt.show()
```

```python
#Correlation Heatmap 
np.random.seed(0)
df=pd.DataFrame(np.random.rand(100,1), columns=['Area 2010-11, Yield 2010-11'])

corr_matrix= df.corr()

plt.figure(figsize=(8,6))
plt.title("Correlation Heatmap", fontsize=16)

sns.heatmap(corr_matrix, annot=True, cmap='winter', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
