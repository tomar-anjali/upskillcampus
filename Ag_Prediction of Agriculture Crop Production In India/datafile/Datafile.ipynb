```python
# setting style
sns.set_style("whitegrid")
 
# plotting data using boxplot for 2006-2008
sns.boxplot(x='2006-07', y='2007-08', data=data)
```

```python
# scatter plot 2006-07 data vs 2007-08 data
plt.scatter(data['2006-07'],data['2007-08'])
 
plt.show()
```

```python
# line plot 2006-07 data vs 2007-08 data
plt.plot(data['2006-07'],data['2007-08'])
 
plt.show()
```

```python
from scipy import stats
```

```python
# assign data
x = data['2010-11']
y = data['2011-12']
 
 
# linear regression 2010-11 data vs 2011-12 data
slope, intercept, r, p, std_err = stats.linregress(x, y)
 
 
# function to return slope
def myfunc(x):
    return slope * x + intercept
 
 
mymodel = list(map(myfunc, x))
 
# scatter
plt.scatter(x, y)
 
# plotting the data
plt.plot(x, mymodel)
 
# display the figure
plt.show()
```
